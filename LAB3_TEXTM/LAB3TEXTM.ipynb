### Validando el usuario mas etiquetado

```python
import re
```

```python
def getAtt(tweet):
  regex =r"@[a-zA-Z0-9]+\s"
  try:
      match = re.findall(regex,tweet)
      match = match[0].replace(' ','')
  except:
      pass
  return match
```

```python
atts = dataset['tweet'].apply(getAtt).dropna()
```

```python
atts.value_counts()
```

```python
mileycyrus = dataset[dataset['tweet'].str.contains("@mileycyrus")]
mileycyrus
```

```python
import spacy
import spacy.cli
```

```python
spacy.cli.download("en_core_web_sm")
spacy.cli.download("en_core_web_md")
```

```python
nlp_en_sm = spacy.load("en_core_web_sm")
nlp_en_sm = spacy.load("en_core_web_md")
```

```python
tamano = mileycyrus['tweet'].str.len()
tamano
```

```python
mileycyruscorpus = pd.concat([mileycyrus,tamano], axis=1)
mileycyruscorpus
```

```python
mileycyruscorpus.set_axis(["fecha","usuario","tweet","largo_tweet"],axis=1, inplace=True)
mileycyruscorpus
```

```python
def normalizando(corpus):
    newCorpus = []
    for doc in corpus:
        newCorpus.append(re.sub(r'[^a-zA-Z0-9\s{1}áéíóúüñÁÉÍÓÚÑ]', '', doc).lower().strip().rstrip('\n').rstrip('\r\n'))
    return newCorpus
```

```python
mileycyruscorpus['tweet'][:7]
```

```python
mileycyruscorpus1 = normalizando(mileycyruscorpus['tweet'])
```

```python
mileycyruscorpus1[:7]
```

```python
def tokenization(corpus):
    newCorpus = []
    for doc in corpus:
        newCorpus.append(nlp_en_sm(doc))
    return newCorpus
```

```python
mileycyruscorpus1 = tokenization(mileycyruscorpus1)
```

```python
mileycyruscorpus1[:7]
```

```python
def removeStops(corpus):
    newCorpus = []
    for doc in corpus:
        s = ''
        for token in doc:
            if(token.is_stop == False):
                s = s + token.text + ' '
        newCorpus.append(s.strip())
    return newCorpus
```

```python
mileycyruscorpus1 = removeStops(mileycyruscorpus1)
```

```python
mileycyruscorpus1[2]
```

```python
def stemming_lemmatization(corpus):
    docs = tokenization(corpus)
    newCorpus = []
    for doc in docs:
        s = ''
        for token in doc:
            s = s + token.lemma_ + ' '
        newCorpus.append(s.strip())
    return newCorpus
```

```python
mileycyruscorpus1 = stemming_lemmatization(mileycyruscorpus1)
```

```python
mileycyruscorpus1[2]
```

```python
def corpus2String(corpus):
    strX = ''
    for doc in corpus:
        strX = strX + doc + ' '
    return strX
```

```python
mileycyruscorpus2 = corpus2String(mileycyruscorpus1)
mileycyruscorpus2
```

```python
mileycyruscorpus3 = mileycyruscorpus2.strip().split(' ')
```

```python
mileycyruscorpus3 =pd.DataFrame(mileycyruscorpus3)
```

```python
mileycyruscorpus3.value_counts(sort = True).head(20)
```

```python
from nltk.stem.snowball import stopwords
from wordcloud import WordCloud,STOPWORDS
stopwords = set(STOPWORDS)
```

```python
stopwords.update(["mileycyrus","miley", "cyrus"])
mileycyruscorpus4 = WordCloud(stopwords = stopwords, max_words = 20, background_color= "white").generate(mileycyruscorpus2)
```
